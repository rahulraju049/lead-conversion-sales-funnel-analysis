Operational Implication:

This represents a direct waste of marketing budget (Total Cost).

1. Sales Representative Conversion Variance:
Our initial aggregation reveals a stark variance in lead conversion efficiency across the sales team. A top-tier segment—consisting of Varsha A.V., Sneha Rajesh, Ashna Sathar, Hamza Anzar, and Viji—consistently maintains an exceptional baseline conversion rate hovering at or above 50%.

Structural Bottlenecks & Root Cause Hypotheses:

Lead Quality Disparity (Routing Bias): The top-performing representatives might be systematically assigned high-intent, warm leads (e.g., direct website sign-ups), while lower-performing tiers are receiving cold or unverified outbound leads. This indicates a potential flaw or lack of blind randomness in our lead distribution algorithm.

Process Drift & Speed-to-Lead: High performers may possess a significantly faster response time ("speed-to-lead"). In digital sales pipelines, contacting a lead within the first 5 minutes increases conversion rates exponentially. Lower conversion tiers may suffer from manual workflows or delayed follow-ups.

Skill Gap or Playbook Deviation: The data suggests that the top 5 reps have mastered a specific objection-handling strategy or pitch rhythm that has not been standardized across the department.

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from catboost import CatBoostClassifier,Pool
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
train_pool=Pool(x_train,y_train,cat_features=cat_features)
test_pool=Pool(x_test,y_test,cat_features=cat_features)
```

```python
parameters={"iterations":[100,150,200],"learning_rate":[0.1,0.3,0.6],
            "depth":[5,8,10],"loss_function":["Logloss"]}
```

```python
grid=GridSearchCV(CatBoostClassifier(verbose=False),parameters,cv=10,scoring="accuracy")
grid.fit(x_train, y_train, cat_features=cat_features)
```

```python
cat=CatBoostClassifier(iterations=200,learning_rate=0.5,depth=5,loss_function="Logloss")
```

```python
cat.fit(train_pool,eval_set=test_pool)
```

```python
y_pred=cat.predict(test_pool)
```

```python
y
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
